# 3D Occupancy Map

```python
import imageio.v2 as imageio
import matplotlib.pyplot as plt

def show_image(path,title):
    image = imageio.imread(path)

    plt.figure(figsize=(8, 8))
    plt.imshow(image)
    plt.axis('off')  
    plt.title(title, fontsize=16)
    plt.show()
```

# NH4 vs Na

```python
show_image('occupancy/nh4cl-occu/new.tga', 'NH4 (Yellow) vs Na (Red) Occupancy Map')
```

# NH4 vs Cl

```python
show_image('occupancy/nh4cl-occu/nh4cl.tga', 'NH4 (Yellow) vs Cl (Blue) Occupancy Map')
```

# Na vs Cl

```python
show_image('occupancy/nacl-occu/nacl.tga', 'Na (Yellow) Cl (Blue) Occupancy Map')
```
