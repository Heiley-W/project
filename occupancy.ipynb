```python
import imageio.v2 as imageio
import matplotlib.pyplot as plt

def show_image(path,title):
    image = imageio.imread(path)

    plt.figure(figsize=(8, 8))
    plt.imshow(image)
    plt.axis('off')  
    plt.title(title, fontsize=16)
    plt.show()
```

```python
show_image('occupancy/image/nh4cl_channel.tga', 'NH4Cl Occupancy Map')
```

```python
show_image('occupancy/image/nacl_channel.tga', 'NaCl Occupancy Map')
```
