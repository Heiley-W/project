## 1. NH4---CLA

```python
!gmx rdf -f ../eq/step5_equilibration.cpt -s ../eq/step5_equilibration.gro -ref "NH4" -sel "CLA" -o rdf_NH4-CLA.xvg
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_settings():
    plt.figure(figsize=(15, 6))
    plt.title("RDF")
    plt.xlabel(r"Position ($\AA$)")
    plt.ylabel('g(r)')
    plt.grid()
    plt.tight_layout()
```

```python
plot_settings()
data = np.loadtxt('rdf_NH4-CLA.xvg', skiprows=25)
pos = data[:, 0]*10
rdf = data[:, 1]
plt.plot(pos,rdf)
plt.show()
```
