## 1. NH4---CLA

```python
!gmx rdf -f nh4cl-1M-box/eq/step5_equilibration.xtc -s nh4cl-1M-box/eq/step5_equilibration.tpr -ref "NH4" -sel "CLA" -o rdf/nh4cl-1M/rdf_NH4-CLA.xvg 
```

```python
!gmx rdf -f nh4cl-0.15M-box/eq/step5_equilibration.xtc -s nh4cl-0.15M-box/eq/step5_equilibration.tpr -ref "NH4" -sel "CLA" -o rdf/nh4cl-0.15M/rdf_NH4-CLA.xvg 
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_settings():
    plt.figure(figsize=(20, 6))
    plt.xlabel(r"Position ($\AA$)")
    plt.ylabel('g(r)')
    plt.grid()
    plt.tight_layout()
```

```python
plot_settings()
data = np.loadtxt('rdf/nh4cl-1M/rdf_NH4-CLA.xvg', skiprows=25)
pos = data[:, 0]*10
rdf = data[:, 1]

data2 = np.loadtxt('rdf/nh4cl-0.15M/rdf_NH4-CLA.xvg', skiprows=25)
pos2 = data2[:, 0]*10
rdf2 = data2[:, 1]

plt.plot(pos,rdf,label="1M NH4Cl")
plt.plot(pos2,rdf2,label="0.15M NH4Cl")
plt.xlim(0, 10)
plt.ylim(0, 35)
plt.title("RDF of NH4+---CLA-")
plt.legend()
plt.show()
```
