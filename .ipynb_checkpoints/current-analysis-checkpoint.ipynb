# 2. Current analysis
## NH4Cl Run 1 to 4

```python
import numpy as np
import matplotlib.pyplot as plt
dt = 5e-11

def plot_settings():
    plt.figure(figsize=(10, 6))
    plt.title("Total Charge Over Time (Run 1 - 4)")
    plt.xlabel("Time (s)")
    plt.ylabel('Total Charge (C)')
    plt.grid()
    plt.tight_layout()
```

```python
runs = ['run1', 'run2', 'run3', 'run4']
species = ['NH4-10', 'CLA-10']
plot_settings()

def load_currents(runs, species, dt):
    currents = {}
    for run in runs:
        for spec in species:
            data = np.loadtxt(f'{run}/current{spec}.dat')
            time = data[:, 0] * dt
            currents[f"{run}_{spec}"] = data[:, 1] * dt
    return currents

currents = load_currents(runs, species, dt)

labels = [
    ('run3_CLA-10', 'Cl- charge (Run 3), 367 events'),
    ('run4_CLA-10', 'Cl- charge (Run 4), 366 events'),
    ('run1_CLA-10', 'Cl- charge (Run 1), 346 events'),
    ('run2_CLA-10', 'Cl- charge (Run 2), 325 events'),
    ('run4_NH4-10', 'NH4+ charge (Run 4), 156 events'),
    ('run3_NH4-10', 'NH4+ charge (Run 3), 124 events'),
    ('run2_NH4-10', 'NH4+ charge (Run 2), 121 events'),
    ('run1_NH4-10', 'NH4+ charge (Run 1), 116 events')
]
time_steps = np.arange(0, len(average_nh4_current) * dt, dt)

for current_key, label in labels:
    plt.plot(time_steps, currents[current_key], label=label)

plt.legend()
plt.show()
```

**Averaging total charges of 4 runs**

```python
nh4_currents = []
cl_currents = []

for run in runs:
    nh4_currents.append(currents[f"{run}_NH4-10"])
    cl_currents.append(currents[f"{run}_CLA-10"])

cl_currents = np.array(cl_currents)
nh4_currents = np.array(nh4_currents)

average_nh4_current = np.mean(nh4_currents, axis=0)
average_cl_current = np.mean(cl_currents, axis=0)

time_steps = np.arange(0, len(average_nh4_current) * dt, dt)

plot_settings()
plt.plot(time_steps, average_cl_current, label='Average Cl- Total Charge')
plt.plot(time_steps, average_nh4_current, label='Average NH4+ Total Charge')
plt.title('Average Total Charges of 4 Runs')
plt.legend()
plt.show()
```

**Compute the average current using linear regression**

$$I(t) = \frac{1}{\Delta t L_z} \sum_{i=1}^{N} q_i [z_i(t + \Delta t) - z_i(t)]$$

To compute an average current at a given bias, instantaneous currents $I(t)$ were first integrated to produce a cumulative current curve. Then, applying a linear regression fit to the cumulative current curve yielded the average current. 

```python
degree = 1

params_nh4, cov_nh4 = np.polyfit(time_steps, average_nh4_current, degree, cov=True)
params_cl, cov_cl = np.polyfit(time_steps, average_cl_current, degree, cov=True)

m_nh4, c_nh4 = params_nh4
perr_nh4 = np.sqrt(np.diag(cov_nh4))
m_cl, c_cl = params_cl
perr_cl = np.sqrt(np.diag(cov_cl))

print("NH4+_________________________________________________")
print("Gradient (m) and intercept (c) of the best-fit line:")
print(f" m = {m_nh4} A ± {perr_nh4[0]} A")  # Error on m
print(f" c = {c_nh4} C ± {perr_nh4[1]} C")  # Error on c

print("Average current:")
print(f" m = {m_nh4} A ± {perr_nh4[0]} A")  # Error on m

print("Cl-_________________________________________________")
print("Gradient (m) and intercept (c) of the best-fit line:")
print(f" m = {m_cl} A ± {perr_cl[0]} A")  # Error on m
print(f" c = {c_cl} C ± {perr_cl[1]} C")  # Error on c

print("Average current:")
print(f" m = {m_cl} A ± {perr_cl[0]} A")  # Error on m
```

```python
x_plot = np.linspace(0,max(time_steps),1001)    
y_plot_nh4 = m_nh4 * x_plot + c_nh4  
y_plot_cl = m_cl * x_plot + c_cl

plot_settings()
plt.plot(time_steps, average_cl_current, label='Cl- Total Charge')
plt.plot(x_plot, y_plot_cl, label='Cl- Best-fit Line')
plt.plot(time_steps, average_nh4_current, label='NH4+ Total Charge')
plt.plot(x_plot, y_plot_nh4, label='NH4+ Best-fit Line')
plt.legend()
```

**Compute the conductance**

The follow equation is used to compute the conductance in (siemens): 
$$ G = \frac{I}{V} $$ 

According to previous calculations, the conductance of Na+ is around 3e-11 S.

```python
voltage = 1.37
conductance_cl = m_cl / voltage
conductance_nh4 = m_nh4 / voltage

print(f"Cl- Conductance = {conductance_cl} S")
print(f"NH4+ Conductance = {conductance_nh4} S")
```

# 3. RMSD analysis
- It has a relatively high rmsd (4 A), compared to the normal 2 A.  

```python
plt.figure(figsize=(15, 6))
plt.grid()
plt.ylabel("RMSD (nm)")
plt.title("RMSD (Run 1 - 4)")

def load_rmsd(runs):
    for run in runs:
        data = np.loadtxt(f'{run}/rmsd-fit.xvg', skiprows=18)
        time = data[:, 0]
        rmsd = data[:, 1]
        plt.plot(time, rmsd, label=f'{run}')
load_rmsd(runs)
plt.legend()
plt.show()
```

# 4. RMSF
- The tail has a high RMSF (0.4 A)
- As the head is fixed, it has low RMSF

```python
def load_rmsf(runs):
    for run in runs:
        plt.figure(figsize=(15, 6))
        plt.grid()
        plt.ylabel("RMSF (nm)")
        plt.title(f"RMSF ({run})")
        data = np.loadtxt(f'{run}/rmsf.xvg', skiprows=17)
        for i in range(0, len(data), 46):
            segment = data[i:i+46]
            resno = segment[:, 0]
            rmsf = segment[:, 1]
            plt.plot(resno, rmsf, label=f'α-helix {i//46 + 1}')
        plt.legend()
        plt.show()
        
load_rmsf(runs)
```

```python
plt.figure(figsize=(15, 6))
plt.grid()
plt.ylabel(r"$R_g$ (nm)")
plt.title("Radius of Gyration (Run 1 - 4)")

def load_rmsd(runs):
    for run in runs:
        data = np.loadtxt(f'{run}/gyrate.xvg', skiprows=27)
        time = data[:, 0]
        rg = data[:, 1]
        plt.plot(time, rg, label=f'{run}')
load_rmsd(runs)
plt.legend()
plt.show()
```
