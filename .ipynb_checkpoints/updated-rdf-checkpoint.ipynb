```python
import numpy as np
import matplotlib.pyplot as plt

def plot_settings():
    plt.figure(figsize=(15,8))
    plt.grid()
    plt.tight_layout()
```

- Na-H, 0.15M vs 1M 
- Na-O, 0.15M vs 1M
- Cl-H, 0.15M vs 1M
- Cl-O, 0.15M vs 1M

## b. Different ions
- Na-H vs Na-O vs Cl-H vs Cl-O (NH4Cl)
- N-H vs N-O vs Cl-H vs Cl-O (NH4Cl)
- Na-H vs Na-O (NaCl) vs N-H vs N-O (NH4Cl)

Reference ([paper](https://www.researchgate.net/figure/The-water-ion-radial-distribution-functions-as-functions-of-the-distance-between-the-ion_fig2_234164234))



```python
data1 = np.loadtxt('rdf/nacl-rdf/1M-run1-rdf/rdf_na-h.xvg',skiprows=25)
x1 = data1[:, 0] 
y1 = data1[:, 1] 

data2 = np.loadtxt('rdf/nacl-rdf/1M-run1-rdf/rdf_na-o.xvg',skiprows=25)
x2 = data2[:, 0] 
y2 = data2[:, 1] 

data3 = np.loadtxt('rdf/nacl-rdf/1M-run1-rdf/rdf_cl-h.xvg',skiprows=25)
x3 = data3[:, 0] 
y3 = data3[:, 1] 

data4 = np.loadtxt('rdf/nacl-rdf/1M-run1-rdf/rdf_cl-o.xvg',skiprows=25)
x4 = data4[:, 0] 
y4 = data4[:, 1] 

plot_settings()
plt.title("Radial Distribution (1M NaCl run 1)")
plt.plot(x1,y1,label = "Na-H")
plt.plot(x2,y2,label = "Na-O")
plt.plot(x3,y3,label = "Cl-H")
plt.plot(x4,y4,label = "Cl-O")
plt.xlabel("Distance (nm)")
plt.ylabel("g(r)")
plt.legend()
plt.show()
```

```python
data1 = np.loadtxt('rdf/nh4cl-rdf/rdf_1M-nh4-h.xvg',skiprows=25)
x1 = data1[:, 0] 
y1 = data1[:, 1] 

data2 = np.loadtxt('rdf/nh4cl-rdf/rdf_1M-nh4-o.xvg',skiprows=25)
x2 = data2[:, 0] 
y2 = data2[:, 1] 

plt.plot(x1,y1,label = "N-H")
plt.plot(x2,y2,label = "N-O")

data1 = np.loadtxt('rdf/nacl-rdf/1M-run1-rdf/rdf_na-h.xvg',skiprows=25)
x1 = data1[:, 0] 
y1 = data1[:, 1] 

data2 = np.loadtxt('rdf/nacl-rdf/1M-run1-rdf/rdf_na-o.xvg',skiprows=25)
x2 = data2[:, 0] 
y2 = data2[:, 1] 


plot_settings()
plt.title("Radial Distribution (1M NH4Cl)")
plt.xlabel("Distance (nm)")
plt.ylabel("g(r)")
plt.legend()
plt.show()
```

# 2. Cumulative number RDF plots

```python
data1 = np.loadtxt('rdf/nacl-rdf/1M-run1-rdf/rdf_na-h_cn.xvg',skiprows=25)
x1 = data1[:, 0] 
y1 = data1[:, 1] 

data2 = np.loadtxt('rdf/nacl-rdf/1M-run1-rdf/rdf_na-o_cn.xvg',skiprows=25)
x2 = data2[:, 0] 
y2 = data2[:, 1] 

data3 = np.loadtxt('rdf/nacl-rdf/1M-run1-rdf/rdf_cl-h_cn.xvg',skiprows=25)
x3 = data3[:, 0] 
y3 = data3[:, 1] 

data4 = np.loadtxt('rdf/nacl-rdf/1M-run1-rdf/rdf_cl-o_cn.xvg',skiprows=25)
x4 = data4[:, 0] 
y4 = data4[:, 1] 

plot_settings()
plt.title("Cumulative Number RDF (1M NaCl run 1)")
plt.plot(x1,y1,label = "Na-H")
plt.plot(x3,y3,label = "Cl-H")
plt.plot(x2,y2,label = "Na-O")
plt.plot(x4,y4,label = "Cl-O")
plt.xlabel("Distance (nm)")
plt.ylabel("Number")
plt.legend()
plt.show()
```
